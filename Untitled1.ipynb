```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
import ternary

# Constants
alphaA = 2.4
alphaB = 1
alphaC = 0.21
N = 10000
x1 = np.array([0.85, 0.12, 0.03])
N1 = x1 * N
m = 10000
n = 38

def montecarlo(N1, alphaA, alphaB, alphaC, m, n):
    A = N1[0]
    B = N1[1]
    C = N1[2]
    points = []

    for i in range(m):
        for j in range(n):
            total = A + B + C
            if total == 0:
                break
            pA = alphaA * A / total
            pB = alphaB * B / total
            pC = alphaC * C / total
            max_p = max(pA, pB, pC)

            random_no=rnd.random()

            if pA / max_p >= random_no:
                A -= 1
            else:
                A = A

            if pB / max_p >= random_no:
                B -= 1
            else:
                B = B

            if pC / max_p >= random_no:
                C -= 1
            else:
                C = C

            A = max(0, A)
            B = max(0, B)
            C = max(0, C)

        total = A + B + C
        if total > 0:
            points.append([A / total, B / total, C / total])

    return points


montecarlo_points = montecarlo(N1, alphaA, alphaB, alphaC, m, n)


def plot_mole_fraction_changes(points):
    points = np.array(points)
    steps = np.arange(len(points))

    plt.figure()
    plt.plot(steps, points[:, 0], label='Benzene (A)')
    plt.plot(steps, points[:, 1], label='Toluene (B)')
    plt.plot(steps, points[:, 2], label='Cumene (C)')
    plt.xlabel('Simulation Step')
    plt.ylabel('Mole Fraction')
    plt.title('Change in Mole Fraction of Each Compound')
    plt.legend()
    plt.show()

plot_mole_fraction_changes(montecarlo_points)
```
