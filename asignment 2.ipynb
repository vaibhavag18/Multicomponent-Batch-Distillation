```python
import numpy as np
import random
import matplotlib.pyplot as plt
relative_volatilities = np.array([2.4, 1, 0.21]) 
N = 10000 
n = 38
m = 220 
mole_fractions = np.zeros((m + 1, 3))
mole_fractions[0] = np.array([0.85, 0.12, 0.03])
molecules = mole_fractions[0] * N
for step in range(1, m+1):
    for _ in range(n):
        p = molecules * relative_volatilities
        max_p = np.max(p)
        r = p / max_p
        random_numbers = np.random.random()
        if(r[0]>random_numbers):
            molecules[0]=molecules[0]-1
        if(r[1]>random_numbers):
            molecules[1]=molecules[1]-1
        if(r[2]>random_numbers):
            molecules[2]=molecules[2]-1
    mole_fractions[step] = molecules / np.sum(molecules)
plt.figure(figsize=(8, 6))
plt.scatter(mole_fractions[:, 1], mole_fractions[:, 2], c=mole_fractions[:, 0], cmap='viridis', s=20)
plt.xlabel('Toluene Mole Fraction')
plt.ylabel('Cumene Mole Fraction')
plt.title('Ternary Diagram')
plt.colorbar(label='Benzene Mole Fraction')
plt.grid(True)
plt.show()
plt.figure(figsize=(8, 6))
plt.plot(range(m + 1), mole_fractions[:, 0], label='Benzene')
plt.plot(range(m + 1), mole_fractions[:, 1], label='Toulene')
plt.plot(range(m + 1), mole_fractions[:, 2], label='Cumene')
plt.xlabel('Simulation Step')
plt.ylabel('Mole Fraction')
plt.title('Change in Mole Fraction')
plt.legend()
plt.grid(True)
plt.show()
```
